```python
import matplotlib.pyplot as plt

# Data
metrics = ['Precision(B)', 'Recall(B)', 'mAP@50(B)', 'mAP@50-95(B)']
values = [0.9126, 0.5068, 0.6538, 0.2845]
colors = ['blue', 'green', 'orange', 'red']

# Create the bar chart
plt.figure(figsize=(6, 4))
bars = plt.bar(metrics, values, color=colors)

# Annotate bars with values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.4f}', ha='center', va='bottom', fontsize=10)

# Title and labels
plt.title('Model Performance Metrics', fontsize=12)
plt.ylabel('Values')

# Show the plot
plt.show()
```
